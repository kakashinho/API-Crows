# 2024

```python
import pandas as pd
import matplotlib.pyplot as plt

exportacoes2024 = 'https://balanca.economia.gov.br/balanca/bd/comexstat-bd/mun/EXP_2024_MUN.csv'
mun = 'https://balanca.economia.gov.br/balanca/bd/tabelas/UF_MUN.csv'
sh4 = 'https://balanca.economia.gov.br/balanca/bd/tabelas/NCM_SH.csv'

df = pd.read_csv(exportacoes2024,sep=';', encoding='latin1')
df_mun = pd.read_csv(mun, sep = ';', encoding='latin1')
df_sh4 = pd.read_csv(sh4, sep=';', encoding='latin1')

display(df)
display(df_mun)
display(df_sh4)
```

##todos os meses

```python
medias_mensais = {}

for mes in range(1,13):
  #filtrando para ser só do estado de São Paulo
  df_sp = df[df['SG_UF_MUN'] == "SP"]

  #filtrando para ser só do mês 1(janeiro)
  df_sp = df_sp[df_sp['CO_MES'] == mes]

  # Tratar casos onde KG_LIQUIDO é zero para evitar divisão por zero
  df_sp = df_sp[df_sp['KG_LIQUIDO']> 0]

  #calculando o valor agregado
  df_sp['VALOR_AGREGADO'] = df_sp['VL_FOB'] / df_sp['KG_LIQUIDO']

  #Agrupar por município e SH4 e calcular a média do valor agregado
  municipios_agregados =df_sp.groupby(['CO_MES','CO_MUN', 'SH4', 'SG_UF_MUN'])['VALOR_AGREGADO'].mean().reset_index()

  #renomear coluna para compatibilidade com a base de municípios
  df_mun = df_mun.rename(columns={'CO_MUN_GEO' : 'CO_MUN'})

  #mesclar com nome do município
  municipios_agregados = municipios_agregados.merge(df_mun, on='CO_MUN', how='left')

  #renomear
  df_sh4 = df_sh4.rename(columns={'CO_SH4' : 'SH4', 'NO_SH4_POR' : 'PRODUTO'})

  municipios_agregados = municipios_agregados.merge(df_sh4, on=['SH4'], how='left')

  #agrupando o nome do municipio e o mes e calculando a média do valor agregado
  municipios_agregados = municipios_agregados.groupby(['NO_MUN_MIN','CO_MES'])['VALOR_AGREGADO'].mean().reset_index()

  # Selecionar colunas relevantes e ordenar
  municipios_top5 = municipios_agregados.sort_values(by='VALOR_AGREGADO', ascending=False).head()
  display(municipios_top5.head())

  #gráfico
  ax = municipios_top5.plot(
      kind='bar',
      title=f'Top 5 valor agregado por municipio mês de {mes}',
      x='NO_MUN_MIN',
      y='VALOR_AGREGADO',
      rot = 45,
      legend=False
  )

  ax.set_xlabel('Municípios')
  ax.set_ylabel('Valor agregado')

#salvando a imagem
  fig = plt.gcf()
  plt.savefig(f'/content/export_2024_mes{mes}.jpg', format='jpg', dpi=300, bbox_inches='tight')

  plt.show()

#guardando a média do valor agregado dos top5 daquele mês
  medias_mensais[mes] = municipios_top5['VALOR_AGREGADO'].mean()

display(medias_mensais)
```
